```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


process_counts = [8, 12, 16, 20, 24, 28]
idle_times_skew = [1186.9898, 1198.1504, 1142.8606, 1138.6444, 1153.764, 1161.3776]
idle_times_uniform = [689.263, 777.5712, 828.0504, 965.8298, 939.1736, 1028.1778]


data = {
    'Processes': process_counts,
    'Skew': idle_times_skew,
    'Uniform': idle_times_uniform
}

df = pd.DataFrame(data)


bar_width = 0.35


r1 = np.arange(len(df['Processes']))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(10, 5))
plt.bar(r2, df['Uniform'], color='tab:green', width=bar_width, edgecolor='grey', label='Uniform')
plt.bar(r1, df['Skew'], color='tab:blue', width=bar_width, edgecolor='grey', label='Skew')


plt.xlabel('Number of Processes', fontsize=20)
plt.ylabel('Total Idle Time (seconds)', fontsize=20)

plt.xticks([r + bar_width/2 for r in range(len(df['Processes']))], df['Processes'], fontsize=20)
plt.yticks(fontsize=20)


plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2, fontsize=16)


for i in range(len(df)):
    plt.text(r1[i] - 0.05, df['Skew'][i] + 10, f"{df['Skew'][i]:.2f}", color='black', fontsize=12)
    plt.text(r2[i] - 0.05, df['Uniform'][i] + 10, f"{df['Uniform'][i]:.2f}", color='black', fontsize=12)


plt.tight_layout(rect=[0, -0.1, 1, 1]) 
plt.savefig("./c3_motivation_idletime.pdf", dpi=300, bbox_inches='tight')
plt.show()
```
